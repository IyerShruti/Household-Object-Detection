```python
import torch
import cv2
from matplotlib import pyplot as plt

# Load the trained model (customize the path to your trained weights)
model = torch.hub.load('ultralytics/yolov5', 'custom', path='runs/train/exp/weights/best.pt', source='local')

def detect_and_visualize(image_path, model):
    # Load image
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB

    # Perform inference
    results = model(img)

    # Results
    results.print()

    # Convert results to numpy array for easier manipulation
    boxes = results.xyxy[0].numpy()  # xyxy format is [x1, y1, x2, y2, confidence, class]
    
    # Plot image with boxes, labels, and confidence
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    
    for box in boxes:
        x1, y1, x2, y2, conf, cls = box
        label = model.names[int(cls)]  # Get the class name using the class index
        color = plt.cm.get_cmap('tab20')(int(cls) % 20)  # Get a color for the class
        rect = plt.Rectangle((x1, y1), x2-x1, y2-y1, linewidth=2, edgecolor=color, facecolor='none')
        plt.gca().add_patch(rect)
        plt.gca().text(x1, y1, f'{label} {conf:.2f}', color='white', size=12, bbox=dict(facecolor=color, alpha=0.75))
    
    plt.show()

# Example usage
image_path = 'path_to_your_test_image.jpg'
detect_and_visualize(image_path, model)
```
